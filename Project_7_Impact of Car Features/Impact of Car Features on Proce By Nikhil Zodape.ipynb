```python
df[df['Engine HP'].isna()]
```

```python
# Define a dictionary with 'Model' as keys and the corresponding 'Engine HP' as values
fill_values = {
    'Model S': 1020,
    'Leaf': 148,
    '500e': 118,
    'Continental': 305,
    'Escape': 181,
    'Fit EV': 123,
    'Focus': 276,
    'Freestar': 201,
    'Impala': 305,
    'Soul EV': 201,
    'MKZ': 188,
    'RAV4 EV': 67,
    'i-MiEV': 63,
    'M-Class': 254
}

# Fill NaN values in 'Engine HP' based on the dictionary
for model, hp_value in fill_values.items():
    df.loc[(df['Model'] == model) & (df['Engine HP'].isna()), 'Engine HP'] = hp_value
```

```python
df[df['Number of Doors'].isna()]
```

```python
# Filling the NaN values of Number of Doors column
df.loc[df.index == 4431, 'Number of Doors'] = 2

# Fill remaining NaN values with 4
df['Number of Doors'].fillna(4, inplace = True)
```
